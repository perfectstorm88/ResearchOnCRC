```python
df1 = df1_[['Symbol','EndDate']] 
df1.head()
code_year_list = []
import re
reg_year = re.compile('\d{4}')
for i in range(1, len(df1)): # 第一行跳过
    code,endDate = df1.iloc[i]['Symbol'], df1.iloc[i]['EndDate']
    if reg_year.match(endDate):
        code_year_list.append(code+'_'+endDate[:4])
    else:
        print(code,endDate)
        
# 构造为map
code_year_map = {}
for k in code_year_list:
    code_year_map[k]= [0,0]
print('code_year_list=',len(code_year_list),len(code_year_map.keys()))
```
